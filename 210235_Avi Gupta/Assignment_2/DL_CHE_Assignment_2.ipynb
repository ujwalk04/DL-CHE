```python
import numpy as np 
import pandas as pd
```

```python
train_data =pd.read_csv("Adsorption_Data.csv")
train_data.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanSquaredError
y = train_data["Adorption Capacity(output)"]
features = [ "C(x1)","H(x2)", "N(x3)","Si(x4)","BET(x5)"]
X = pd.get_dummies(train_data[features])
print(X)
```

```python
X_test = pd.get_dummies(train_data[features])
model = Sequential(
[ Dense(128, activation = 'relu'),
Dense(64, activation = 'relu'),
Dense(1, activation = 'linear') # < softmax activation here
]
)
model.compile(
loss=tf.keras.losses.MeanSquaredError(),
optimizer=tf.keras.optimizers.Adam(0.001),
)
model.fit(
X,y,
epochs=1000
)
predictions = model.predict(X)
```

```python
print(predictions)
print(y)
```

```python
import matplotlib.pyplot as plt
x=np.arange(35)
plt.plot(x,y,'g',x,predictions,'r')
plt.savefig('my_image.png')
plt.show()
```
